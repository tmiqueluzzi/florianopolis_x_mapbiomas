```python
import os
from PIL import Image

# Caminhos base
PASTA_IMAGENS = "Imagens"
PASTA_GIFS = "GIFs"

# Cria a pasta de GIFs se não existir
os.makedirs(PASTA_GIFS, exist_ok=True)

# Extensões de imagem aceitas
EXTENSOES = (".jpg", ".jpeg", ".png", ".webp")

# Percorre cada distrito
for distrito in os.listdir(PASTA_IMAGENS):
    caminho_distrito = os.path.join(PASTA_IMAGENS, distrito)

    if not os.path.isdir(caminho_distrito):
        continue

    # Lista e ordena as imagens
    imagens = sorted([
        os.path.join(caminho_distrito, img)
        for img in os.listdir(caminho_distrito)
        if img.lower().endswith(EXTENSOES)
    ])

    if not imagens:
        print(f"⚠️ Nenhuma imagem em {distrito}")
        continue

    frames = []
    for img_path in imagens:
        img = Image.open(img_path).convert("RGB")
        frames.append(img)

    # Caminho do GIF final
    caminho_gif = os.path.join(PASTA_GIFS, f"{distrito}.gif")

    # Salva o GIF (500 ms = 0.5s por frame)
    frames[0].save(
        caminho_gif,
        save_all=True,
        append_images=frames[1:],
        duration=500,
        loop=0
    )

    print(f"✅ GIF criado: {caminho_gif}")
```
